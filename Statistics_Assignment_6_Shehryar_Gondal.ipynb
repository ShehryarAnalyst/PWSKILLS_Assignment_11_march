__Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test ?__

__Ans)__ A t-test and a z-test are statistical hypothesis tests used to determine if there is a significant difference between the means of two groups or if a single sample mean is significantly different from a population mean.

The main difference between a t-test and a z-test is the use of the sample size and the population standard deviation. A t-test is appropriate when the sample size is small (typically less than 30) and the population standard deviation is unknown. In contrast, a z-test is used when the sample size is large (typically greater than 30) and the population standard deviation is known.

__An example scenario where a t-test would be appropriate is when a researcher wants to test if there is a significant difference in the mean weight of two different groups of mice, where the sample size for each group is less than 30. In this case, a two-sample t-test would be used.__

__An example scenario where a z-test would be appropriate is when a researcher wants to test if the mean height of a population of students is significantly different from the mean height of a known population of students, where the sample size is greater than 30. In this case, a one-sample z-test would be used.__

__Q2 Differentiate between one-tailed and two-tailed tests ?__

__Ans)__ In hypothesis testing, a statistical test can be either one-tailed or two-tailed.

A one-tailed test is a statistical test in which the null hypothesis is rejected if the test statistic falls either entirely in the upper or lower tail of the probability distribution. The alternative hypothesis is only interested in detecting an effect in one direction. For example, if a researcher wants to test if a new medication will increase the response time of participants, a one-tailed test will be used in which the null hypothesis would be that there is no difference between the groups and the alternative hypothesis would be that the medication will increase response time. The rejection region will only be on one side of the distribution (either the upper or lower tail) depending on the direction of the alternative hypothesis.

In contrast, a two-tailed test is a statistical test in which the null hypothesis is rejected if the test statistic falls in either tail of the probability distribution. The alternative hypothesis is interested in detecting any difference from the null hypothesis, regardless of the direction. For example, if a researcher wants to test if a new treatment for a disease is effective, a two-tailed test will be used, in which the null hypothesis would be that there is no difference between the groups, and the alternative hypothesis would be that there is a difference (either an increase or decrease in effectiveness). The rejection region will be on both sides of the distribution (both the upper and lower tail).

__In summary, the difference between one-tailed and two-tailed tests lies in the directionality of the alternative hypothesis and the location of the rejection region in the probability distribution.__

__Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error ?__

__Ans)__ In hypothesis testing, there are two types of errors that can occur: Type 1 error and Type 2 error.

Type 1 error occurs when the null hypothesis is rejected when it is actually true. In other words, a Type 1 error is a false positive, where the researcher concludes that there is a significant difference or effect when there is not. The probability of making a Type 1 error is denoted by alpha (α) and is set by the researcher before conducting the test.

An example scenario of a Type 1 error is a medical researcher testing a new drug to lower blood pressure. If the researcher concludes that the drug is effective in lowering blood pressure when it is not, this would be a Type 1 error.

Type 2 error occurs when the null hypothesis is not rejected when it is actually false. In other words, a Type 2 error is a false negative, where the researcher fails to detect a significant difference or effect when there is one. The probability of making a Type 2 error is denoted by beta (β) and is influenced by factors such as sample size, effect size, and statistical power.

An example scenario of a Type 2 error is a medical researcher testing a new drug to lower blood pressure. If the researcher concludes that the drug is not effective in lowering blood pressure when it is, this would be a Type 2 error.

In summary, Type 1 error is the probability of rejecting a true null hypothesis, while Type 2 error is the probability of not rejecting a false null hypothesis. It is important to minimize both types of errors in hypothesis testing to ensure accurate and reliable results.

__Q4: Explain Bayes's theorem with an example__ 

__Ans)__ Bayes's theorem is a statistical concept that provides a way of updating probabilities based on new evidence. It is named after Reverend Thomas Bayes, an 18th-century British statistician and theologian who first formulated the theorem.

Bayes's theorem is based on the following formula:

P(A|B) = P(B|A) * P(A) / P(B)

where:

* P(A|B) is the probability of A given B
* P(B|A) is the probability of B given A
* P(A) is the prior probability of A
* P(B) is the prior probability of B

__In other words, Bayes's theorem allows us to calculate the probability of an event A happening given that event B has already occurred, by taking into account the prior probabilities of A and B and the likelihood of B given A.__


__Here is an example to illustrate Bayes's theorem:__

Suppose a medical test has a 99% accuracy rate in detecting a rare disease, which affects only 0.1% of the population. If a person tests positive for the disease, what is the probability that they actually have the disease?

Let's use Bayes's theorem to calculate this probability. Let A be the event of having the disease and B be the event of testing positive for the disease. Then we have:

* P(A) = 0.001 (prior probability of having the disease)
* P(B|A) = 0.99 (probability of testing positive given that the person has the disease)
* P(B|~A) = 0.01 (probability of testing positive given that the person does not have the disease)
* P(~A) = 0.999 (prior probability of not having the disease)

Using these probabilities in Bayes's theorem, we get:

* P(A|B) = P(B|A) * P(A) / P(B|A) * P(A) + P(B|~A) * P(~A)
* P(A|B) = 0.99 * 0.001 / (0.99 * 0.001 + 0.01 * 0.999)
* P(A|B) = 0.09

__Therefore, the probability that a person actually has the disease given that they tested positive for it is only 9%. This means that even if a person tests positive for the disease, there is still a 91% chance that they do not have it, due to the low prevalence of the disease in the population__

__Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example ? __ 

__Ans)__ A confidence interval __is a range of values that is likely to contain the true population parameter with a specified level of confidence.__ In other words, it is an estimate of the range of values that is likely to contain the true value of a population parameter, such as the population mean or population proportion.

The calculation of a confidence interval depends on the sample size, sample mean or proportion, and the level of confidence chosen. The level of confidence is typically set at 95%, meaning that there is a 95% chance that the true population parameter lies within the calculated interval. The formula for calculating a confidence interval varies depending on the type of parameter being estimated (e.g., mean, proportion) and the distribution of the sample data.

__Here is an example of how to calculate a confidence interval for the population mean:__

Suppose we want to estimate the average height of all students in a university. We randomly sample 100 students and measure their heights, obtaining a sample mean of 68 inches and a standard deviation of 3 inches. We want to calculate a 95% confidence interval for the population mean height.

Using the formula for a confidence interval for the population mean, we have:

* Confidence interval = sample mean +/- (critical value * standard error)

where the critical value is obtained from a t-distribution table based on the degrees of freedom (df = n - 1), and the standard error is the standard deviation divided by the square root of the sample size:

* Critical value for a 95% confidence interval with 99 degrees of freedom = 1.984 (obtained from a t-distribution table)

* Standard error = 3 / sqrt(100) = 0.3

* Substituting these values into the formula, we get:

* Confidence interval = 68 +/- (1.984 * 0.3)
* Confidence interval = (67.4, 68.6)

__Therefore, we can say with 95% confidence that the true population mean height lies between 67.4 and 68.6 inches.__

__Solution:__

Let's use Bayes' Theorem to calculate the probability that the defective product is a type A product. Let A be the event that the product is a type A product, and B be the event that the product is defective. We know the prior probabilities of these events, as well as the conditional probabilities:

* P(A) = 0.5 (since there are equal numbers of type A and type B products produced)
* P(~A) = 0.5 (since there are equal numbers of type A and type B products produced)
* P(B|A) = 0.05 (the probability that a type A product is defective)
* P(B|~A) = 0.02 (the probability that a type B product is defective)

Using Bayes' Theorem, we have:

* P(A|B) = P(B|A) * P(A) / [P(B|A) * P(A) + P(B|~A) * P(~A)]

Plugging in the given values, we get:

* P(A|B) = 0.05 * 0.5 / [0.05 * 0.5 + 0.02 * 0.5]
* P(A|B) = 0.71 or 71%
